```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.common.exceptions import TimeoutException
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC
from time import sleep
import pandas as pd
```

```python
ff_options = webdriver.ChromeOptions()
base_url = 'https://shopee.sg/Local-Ready-Stock-3-PLY-Disposable-Face-Masks-50-pcs-Adult-Fastest-Shipping-i.170474682.7040086717'

ff_options.add_argument('disable-notifications')
ff_options.add_argument('--disable-infobars')
ff_options.add_argument('--start-maximized')
ff_options.add_argument('--disable-extensions')
ff_options.add_argument('--headless')
```

```python
driver = webdriver.Chrome(executable_path = r'C:\Program Files\Google\Chrome\Application\chromedriver.exe',options = ff_options)
driver.get(base_url)
sleep(3)`
```

```python
## Method 1: Get all info one by one.
users = []
ratings = []
variations = []
reviews = []
dates = []

driver.execute_script("window.scrollTo(0, window.scrollY + 2000);")
sleep(2)

while True:
    try:
        current_page = driver.find_elements_by_css_selector('.shopee-button-solid--primary')[0].text
        print(f"Working on page {current_page}")
        
        for item in driver.find_elements_by_css_selector('.shopee-product-rating__main'):
            user = item.find_elements_by_css_selector('.shopee-product-rating__author-name')
            variation = item.find_elements_by_css_selector('.shopee-product-rating__variation')
            review = item.find_elements_by_css_selector('.shopee-product-rating__content')
            rating = item.find_elements_by_css_selector('.icon-rating-solid')
            date = item.find_elements_by_css_selector('.shopee-product-rating__time')
            
            users.append(user[0].text if len(user)!=0 else "")
            variations.append(variation[0].text if len(variation)!=0 else "")
            reviews.append(review[0].text if len(review)!=0 else "")
            ratings.append(len(rating))
            dates.append(date[0].text if len(date)!=0 else "")

        driver.find_elements_by_css_selector('.shopee-icon-button--right')[0].click()
        sleep(1)
        new_page = driver.find_elements_by_css_selector('.shopee-button-solid--primary')[0].text
        if new_page == current_page:
            print(f"Last page arrived at page {new_page}.")
            break
    except TimeoutException:
        print ("Loading took too much time!-Try again") 
```

```python
shopee_data = pd.DataFrame({"user":users, "variation":variations, "rating":ratings, "review": reviews, "date":dates})
shopee_data.tail()
```

```python
shopee_data.to_csv('shopee_mask_reviews.csv')
```
